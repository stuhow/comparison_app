```python
import pandas as pd
carrier = []
price = []
stops = []
origin = []
destination = []
departure = []
arrival = []
duration =[]

for i in range(len(response['data'])):
    carrier.append(response['data'][i]['legs'][0]['carriers'][0]['name'])
    price.append(response['data'][i]['price']['amount'])
    stops.append(response['data'][i]['legs'][0]['stop_count'])
    origin.append(response['data'][i]['legs'][0]['origin']['name'])
    destination.append(response['data'][i]['legs'][0]['destination']['name'])
    departure.append(response['data'][i]['legs'][0]['departure'])
    arrival.append(response['data'][i]['legs'][0]['arrival'])
    duration.append(response['data'][i]['legs'][0]['duration'])
```

```python
data = {'carrier': carrier,
        'price': price,
        'stops': stops,
        'origin': origin,
        'destination': destination,
        'departure': departure,
        'arrival': arrival,
        'duration': duration
       }
df = pd.DataFrame.from_dict(data)
```
